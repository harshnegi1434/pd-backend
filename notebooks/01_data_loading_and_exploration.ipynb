# PhishDetective : Phishing Detection System - Data Loading And Exploration

```python
# Importing Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
```

## 1. Loading Data:

The dataset is borrowed from Kaggle, [Phishing Website Detector Dataset](https://www.kaggle.com/eswarchandt/phishing-website-detector).

A collection of website URLs for 11000+ websites. Each sample has 30 website parameters and a class label identifying it as a phishing website or not (1 or -1).

The overview of this dataset is, it has 11054 samples with 32 features. Download the dataset from the [link provided](https://www.kaggle.com/eswarchandt/phishing-website-detector).


```python
# Loading data into dataframe

data = pd.read_csv("../data/raw/phishing_data.csv")
data.head()
```

## 3. Visualizing the data:
Few plots and graphs are displayed to find how the data is distributed and the how features are related to each other.

```python
# Correlation heatmap

plt.figure(figsize=(20,20))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
# Pairplot for particular features

df = data[['PrefixSuffix-', 'SubDomains', 'HTTPS', 'AnchorURL', 'WebsiteTraffic', 'class']]
sns.pairplot(data = df, hue="class", corner=True)
```

```python
# Phishing count in pie chart

data['class'].value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.title("Phishing Count")
plt.show()
```

## 4. Exporting the processed dataset:
After completing the Exploratory Data Analysis and little data pre-processing, exporting the processed dataset into csv file for further use for model training

```python
# Exporting the dataset

output = "../data/processed/phishing_processed.csv"

data.to_csv(output, index=False)

print("Preprocessed Data Exported Succesfully!")
```
